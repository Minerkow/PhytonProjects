```python
data = np.loadtxt("2_.dat")
x = 0.00025*data[:, 0] - 45*0.00025
y = np.sqrt(2*((data[:, 1] - 550)/8.663)/1.2)
y1 = np.array([400, 400])
x1 = np.array([0, 0.0275])
```

```python
neg = []
pos = []
for i in x:
    # если элемент меньше нуля,
    if i < 0:
        # то добавляется к списку отрицательных чисел
        neg.append(i)
    # если элемент больше нуля,
    elif i >= 0:
        # то добавляется к списку положительных чисел
        pos.append(i)
    # Примечание. Ноль не добавляется никуда
neg = np.array(neg)
pos = np.array(pos)
```

```python
plt.figure(figsize=(15,8))
ax = plt.axes()
ax.grid(color='grey', linestyle='-', linewidth=0.5)
ax.set_facecolor("white")
#ax.set(option='on')
#ax.set_facecolor("w")
plt.title("График скорости для сечения на расстоянии 0мм")
plt.xlabel("Координата, м")
plt.ylabel("Скорость, м/c")
#plt.scatter(x, y, color="b")
#plt.plot(x, y, color='b')
plt.plot(x, y, color='black', linestyle='-', marker = '.',
     markerfacecolor='green', markersize=9, linewidth=1)
ax.spines['left'].set_position('zero')
ax.spines['left'].set_color("black")
ax.spines['bottom'].set_color('black')
#ax.xaxis.set_ticklabels([500], color = 'k')
#plt.plot(x1, y1, color='r', label="0 Па")
#plt.plot(x2, y2, color='r')
plt.xlim(-0.012, 0.017)
#ax.fill_between(pos, y[44:108], 1)
plt.grid(True, )
plt.savefig("4_gr")
```

```python
s1 = 0
s2 = 0
for i in range(pos.size - 1):
    s1 = s1 + ((y[44 + i] + y[44 + i + 1])/2) * 0.00025
print(s1)
for i in range(neg.size - 1):
    s2 = s2 + ((y[i] + y[i + 1])/2) * 0.00025
print(s2)
```
